```python
import pandas as pd

df = pd.read_csv("../data/raw/Unemployment in India.csv")
df.head()
```

```python
df.columns = [
    col.strip()
       .lower()
       .replace(" ", "_")
       .replace("(", "")
       .replace(")", "")
       .replace("%", "percent")
    for col in df.columns
]

df.columns
```

```python
df["date"] = pd.to_datetime(df["date"])
df.dtypes
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.shape
```

```python
df.to_csv("../data/processed/unemployment_cleaned.csv", index=False)
```
