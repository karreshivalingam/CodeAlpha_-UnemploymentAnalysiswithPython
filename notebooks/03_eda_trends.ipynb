```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/processed/unemployment_cleaned.csv")
df.head()
```

```python
monthly_trend = (
    df.groupby("date")["estimated_unemployment_rate_percent"]
      .mean()
      .reset_index()
)
monthly_trend.head()
```

```python
plt.figure(figsize=(10,5))
plt.plot(
    monthly_trend["date"],
    monthly_trend["estimated_unemployment_rate_percent"]
)

plt.title("Average Unemployment Rate Over Time (India)")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.grid(True)
plt.show()
```
